```python
#2.For the above matrices ùë® and ùë©, compute their Frobenius norms (You cannot use the Python built-in function).

def Frobeniusfunc(A):
    F = 0
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            F += np.square(A[i][j])
#             print(A[i][j])
    return np.sqrt(F)
print('Frobenius norms A =', Frobeniusfunc(A))
print('Frobenius norms B =', Frobeniusfunc(B))
```

```python
#3.Load the data ‚ÄúKFold.xlsx‚Äù and perform 5-fold cross-validation, then compute the average accuracy. 
#Assume the accuracy from the first fold to the fifth fold respectively is 90%, 80%, 75%, 82%, and 95%. 
 
import pandas as pd
import numpy as np
import time

names = ['Sample','var1','var2','var3','var4','var5','var6','var7','var8']
data = pd.read_excel('KFoldCV.xlsx', names = names)
A = [90, 80, 75, 82, 95]

class KFold:
    def __init__(self,n_folds,shuffle):
        self.n_folds = n_folds
        self.shuffle = shuffle
    def split(self,data):
        #Each fold is then used a validation set once while the k - 1 remaining fold form the training set.
        #The first n % n_folds folds have size n // n_folds + 1, other folds have size n // n_folds.
        n = len(data)
        remain = len(data) % self.n_folds
        x = np.arange(n)
        if self.shuffle == True:
            np.random.shuffle(x)
        for i in range (0, self.n_folds):
            start = 0
            end = 0
            if i < remain:
                start = (n // self.n_folds + 1) * (i)
                end = start + (n // self.n_folds + 1) 
            else:
                start = (n // self.n_folds + 1) * remain + (n // self.n_folds) * (i - remain)
                end = start + (n // self.n_folds)
            temp = np.split(x,[start,end])
            train_index = np.append(temp[0],temp[2])
            test_index = temp[1]
            yield np.sort(train_index), np.sort(test_index)
kf = KFold(n_folds=5,shuffle = True)
count = 0
for train_index, test_index in kf.split(data):
    print("TRAIN:", train_index, "TEST:", test_index)
    print("X_train", data.iloc[train_index][names[1:]])
    print("Y_train", data.iloc[train_index]['Sample'])
    print("X_test", data.iloc[test_index][names[1:]])
    print("Y_test", data.iloc[test_index]['Sample'])
    print("Accurancy: ", A[count], "%")
    count = count + 1
print("Average Accurancy: ", np.average(A), "%")
```
